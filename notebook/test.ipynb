```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('..')

from backtest.data_handler import DataHandler
```

```python
symbols = {'BTCUSDT','ETHUSDT','BNBUSDT'}
# load data from csv
data_dict = {
    'close': pd.DataFrame(),
    'high': pd.DataFrame(),
    'low': pd.DataFrame(),
    'open': pd.DataFrame(),
    'volume': pd.DataFrame()
}

for symbol in symbols:
    file_path = f'D:/NYCU_studygroup_system/data/pricevolumn/{symbol}.csv'
    data = pd.read_csv(file_path)
    
    close_df = data[['timestamp', 'close']]
    close_df.set_index('timestamp', inplace=True)
    close_df.columns = [symbol]
    if data_dict['close'].empty:
        data_dict['close'] = close_df
    else:
        data_dict['close'] = pd.concat([data_dict['close'], close_df], axis=1)
    
    high_df = data[['timestamp', 'high']]
    high_df.set_index('timestamp', inplace=True)
    high_df.columns = [symbol]
    if data_dict['high'].empty:
        data_dict['high'] = high_df
    else:
        data_dict['high'] = pd.concat([data_dict['high'], high_df], axis=1)

    low_df = data[['timestamp', 'low']]
    low_df.set_index('timestamp', inplace=True)
    low_df.columns = [symbol]
    if data_dict['low'].empty:
        data_dict['low'] = low_df
    else:
        data_dict['low'] = pd.concat([data_dict['low'], low_df], axis=1)

    open_df = data[['timestamp', 'open']]
    open_df.set_index('timestamp', inplace=True)
    open_df.columns = [symbol]
    if data_dict['open'].empty:
        data_dict['open'] = open_df
    else:
        data_dict['open'] = pd.concat([data_dict['open'], open_df], axis=1)

    volume_df = data[['timestamp', 'volume']]
    volume_df.set_index('timestamp', inplace=True)
    volume_df.columns = [symbol]
    if data_dict['volume'].empty:
        data_dict['volume'] = volume_df
    else:
        data_dict['volume'] = pd.concat([data_dict['volume'], volume_df], axis=1)

print(data_dict)
```

```python
handler = DataHandler(data_dict)

close = handler['close']
print(close)

open_data = handler['open']
print(open_data)
```
